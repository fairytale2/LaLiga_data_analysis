```python
import pandas as pd
```

```python
pl = pd.read_csv('players.csv')
```

<b> Before performing any analysis on the players' stats dataset, it is important to clean the data to ensure its quality and consistency </b>

```python
pl.head()
```

### <font color = 'red'>Here is a breakdown of the information each column represents:</font><br>

<b>Player Name: The name of the player.<br>
Team-name: The name of the team the player belongs to.<br>
Age: The age of the player.<br>
Position: The position played by the player.<br>
App: The number of appearances (matches) the player has made.<br>
MinP: Minutes played by the player.<br>
Goals: The number of goals scored by the player.<br>
Assist: The number of assists made by the player.<br>
YC: The number of yellow cards received by the player.<br>
RC: The number of red cards received by the player.<br>
SPG: Shots per game by the player.<br>

```python
# Quick overview of important statistical measures
pl.describe()

# can help in identifying missing values. 
# If the count is lower than the total number of observations
# It helps identifying potential issues with the data
```

```python
# Check for any missing values
pl.isnull().sum()
```

```python
pl.head()
```

```python
# Should fill the null values by 0
# 0 is the common approach to replace null values
pl.fillna(0, inplace=True)
```

```python
pl.head()
# Null values replaced by 0
```

```python
# Check for any duplicate rows in the dataset based on the player name
pl.duplicated(subset='Player Name').sum()
```

```python
# View the duplicated values 
duplicated_player = pl[pl.duplicated(subset='Player Name', keep=False)]
duplicated_player
```

```python
# Remove the duplicated values
pl.drop_duplicates(subset='Player Name', keep='last', inplace=True) 
```

```python
pl.duplicated(subset='Player Name').sum()
```

```python
pl['App']
# The App column indicates values between parenthesis which means the number of substitute appearances made
# No need for thoose values in the analysis process
```

```python
pl['App'] = pl['App'].str.replace(r'\(\d+\)', '', regex=True)
pl['App']
```

```python
pl[pl['App'] == '0']
```

<b> Players with no appearances have some valid numbers, <br>
    which contradicts the reality.<br>
    We should remove them</b>

```python
# Remove rows with 0 values
pl = pl[pl['App'] != '0']
# check if they removed correctly
pl[pl['App'] == '0']
# Rows removed successfully
```

```python
# Check the column's type
pl.dtypes
```

<b> ['App', 'Goals', 'Assist', 'YC', 'RC', 'MOTM'] <br> Thoose columns should ben integer</b>

```python
columns = ['Goals', 'Assist', 'YC', 'RC', 'MOTM'] 

pl.loc[:, columns] = pl.loc[:, columns].astype(int)
```

```python
# pl['App'] = pl['App'].astype(int)
# ValueError: invalid literal for int() with base 10: '22()'
```

```python
pl['App'] = pl['App'].str.replace(r'\(.*\)', '',  regex=True)
```

```python
# still exist none values we should replace them with '0'
pl.fillna('0', inplace= True)
```

```python
# Convert to integer
pl['App'] = pl['App'].astype(int)
```

```python
pl['App'].dtypes
# The variable type become integer
```

```python
# Check if there are any null values
pl.isnull().sum()
```

```python
# Check the types of the data
pl.dtypes
```

```python
# Focus on 'PS%' column which is an str column
# Check if the column contain only float value
pattern = r'^\d*\.?\d+$'  # Regular expression pattern to match float numbers
x = pl['PS%'].str.contains(pattern, regex=True, na=False) == False
pl[x]
```

```python
# Replace the value
pl['PS%'] = pl['PS%'].str.replace('-', '0')
```

```python
# Check if the row cleaned correctly
pl['PS%'].str.contains('-').sum()
```

```python
# Change the column type
pl['PS%'] = pl['PS%'].astype(float)
```
